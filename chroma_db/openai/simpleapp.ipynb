```python
## Data Ingestion -- From the website we need to scrape the data

from langchain_community.document_loaders import WebBaseLoader
```

```python
loader = WebBaseLoader("https://docs.langchain.com/langsmith/observability-concepts")
loader
```

```python
docs = loader.load()
docs
```

```python
### Load Data ---> Docs ---> Divide our text into chunks --->text ---> vectors ---> Vector Embeddings ---> Vector Store DB 

from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000 , chunk_overlap=200)
documents = text_splitter.split_documents(docs)
documents
```

```python
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
)
 
```

```python
vectorstoredb = FAISS.from_documents(documents=docs , embedding=embeddings)
vectorstoredb
```

```python
query = "project is a container for all the traces related to a single application or service."
result = vectorstoredb.similarity_search(query)
result[0].page_content
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ['Gemini_API_Key'] = os.getenv('Gemini_API_Key')
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2'] = 'true'
os.environ["LANGCHAIN_PROJEC"] = os.getenv("LANGCHAIN_PROJEC")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.environ['Gemini_API_Key']
)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
prompts= ChatPromptTemplate.from_template(
    """
Answer the following question based only on the provided context:
<context>
{context}
</context>
"""
)

document_chain = create_stuff_documents_chain(llm , prompt=prompts)
```

```python
document_chain
```

```python
from langchain_core.documents import Document
document_chain.invoke({
    "input":"thread is a sequence of traces representing a single conversation. Each turn in a multi-turn conversation is its own trace, but traces are linked by a shared identifier",
    "context":[Document(page_content="thread is a sequence of traces representing a single conversation. Each turn in a multi-turn conversation is its own trace, but traces are linked by a shared identifier")]
})
```

However , we want the documents to fisrt come from the retriver we just set up . That way we can use the retriver to dynamic select the most relevant document and pass those in the for a given question 


```python
## retriver ===> vectorstoredb 
vectorstoredb
```

```python
retriver = vectorstoredb.as_retriever()
from langchain.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriver , document_chain)
retrieval_chain
```

```python
## Get the response from the LLM
response = retrieval_chain.invoke({"input":"thread is a sequence of traces representing a single conversation. Each turn in a multi-turn conversation is its own trace, but traces are linked by a shared identifier"})
response['answer']
```
