HOW TO SPLIT BY HTML HEADER
HTML HEADER TEXT SPLITTER is a "Structure-aware" chunker that splits text at the HTML element level and adds metadata for each header "relevant" to any given chunk. It can return chunk element by elemenr or cobine elements with the same metadata , with the objective of (a) keeping related text grouped (more or less)  semantically and (b) preserving context-rich information encodeed in documents structure . It can be used with other text splitters as part of a chunking pipeline.


```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <h1>Main Title</h1>
    <p>This is the introduction.</p>

    <h2>Section 1</h2>
    <p>Content of section 1.</p>

    <h2>Section 2</h2>
    <p>Content of section 2.</p>
</body>
</html>
"""

header_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]


html_splitter = HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits


 
```

```python
import requests
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://socialsrecovery.com/"

response = requests.get(url)
html_content = response.text

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

html_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

html_header_splits = html_splitter.split_text(html_content)

print(len(html_header_splits))
print(html_header_splits[0])
```
